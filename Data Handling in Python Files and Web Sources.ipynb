```python
import requests

api_url = "https://jian.sh/malaysia-api/state/v1/"
state_name = input("Enter the name of a state: ")

response = requests.get(api_url + f"{state_name}.json")
if response.status_code == 200:
    state_data = response.json()
    population = state_data['population']
    print(f"Population of {state_name}: {population}")
else:
    print(f"{state_name} is not found in the data.")

    
```

```python
import requests

api_url = "https://jian.sh/malaysia-api/state/v1/all.json"
response = requests.get(api_url)
malaysia_state_data = response.json()

while True:
    state = input("Enter the name of a state: ")
    if state in malaysia_state_data:
        population = malaysia_state_data[state]['population']
        print(f"Population of {state}: {population}")
    else:
        print(f"{state} is not found in the data.")
    another_query = input("Do you want to check another state? (yes/no): ")
    if another_query.lower() != 'yes':
        break

        
```
